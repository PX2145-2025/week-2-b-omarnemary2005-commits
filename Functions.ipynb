## More about functions
Sometimes, there can be lots of variables that remain the same when calling a function. If that’s the case you can simplify the syntax a bit. For example, let’s write a function to find how long it takes a mass to hit the ground when dropped (from rest) from a height h. The function will look like:

````python
def T(h, g):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above the Earth's surface. g in ms^-2.
    '''
    return sqrt(2*h/g)
````

This is called as follows:

````python
height = 1  # metre
print ('Time from', height, 'metre', 'is' , T(height, 9.8),'s')
````

You *could* program this with a hard-coded value for g, to make the function call simpler:

````python
def T(h):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above the Earth's surface.
    '''
    g=9.81 # in ms^-2
    return sqrt(2*h/g)

height = 1 # metre
print ('Time from', height, 'metre', 'is' , T(height),'s')
````

but then of course it wouldn’t work for Mars’ gravity. A better way is

````python
def T(h, g=9.81):
    '''
    Returns fall time [s] of a mass released from a height
    h [m] above a planet's surface. Default g is 9.81 ms^-2
    '''
    return sqrt(2*h/g)

height = 1  # metre
print('Time from', height, 'metre', 'on Earth is' , T(height),'s')
print('Time from', height, 'metre', 'on Mars is' , T(height, g=3.7),'s')
````

````python
for height in [1,2,5,10]:
    print('Time from', height, 'metre', 'on Earth is' , T(height),'s')
````

[0] Write a function `convert(angle)` to convert angles in degrees to radians. Write another function `recenter(angle)` to wrap angles in radian into the $(-\pi,\pi]$ interval. Save those two functions into a file named `anglesutils.py`.

A centrifuge's loading position at $t=0\,\text{s}$ is $45^\circ$ degrees, and start accelerating at $\alpha=2^{\circ}\,\text{s}^{-2}$. So it's position in degrees is given as a function of time (in seconds) by:

$$
position(t)=45+\frac{1}{2}\,2\,t^2
$$

Use the `anglesutils` module you just created to plot that function from $t=0\,\text{s}$ to $t=30\,\text{s}$ in radian in the $(-\pi,\pi]$ interval.

(this exercise is for demonstration purposes and won't be marked)

````python
%%writefile anglesutils.py
from numpy import deg2rad, pi

def convert(angle_deg):
    """Convert degrees to radians."""
    return deg2rad(angle_deg)

def recenter(angle_rad):
    """Wrap radians into (-π, π]."""
    return (angle_rad + pi) % (2 * pi) - pi
````

````python
from anglesutils import convert, recenter
from numpy import linspace
import matplotlib.pyplot as plt

# Parameters
alpha = 2  # angular acceleration in deg/s^2
t = linspace(0, 30, 500)  # time from 0s to 30s

# Position in degrees
position_deg = 45 + 0.5 * alpha * t**2

# Convert to radians
position_rad = convert(position_deg)

# Recenter into (-π, π]
position_recentered = recenter(position_rad)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, position_recentered, label="Centrifuge position")
plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.title("Centrifuge position vs. time (wrapped to (-π, π])")
plt.legend()
plt.grid(True)
plt.show()
````

## Exercise 1

[4] Define a function `func1(time,amplitude,frequency,phase)` which returns a sine wave (as a function of time) of chosen amplitude, frequency and phase. Test it out with different amplitude, frequency and phase values by plotting suitable graphs. Set the default values for ampliture, frequency and phase to be 1, 1 and 0, respectively.

````python
from numpy import sin, pi

def func1(time, amplitude=1, frequency=1, phase=0):
    """
    Generate a sine wave.

    Parameters:
        time (array): time values
        amplitude (float): amplitude of the wave (default 1)
        frequency (float): frequency in Hz (default 1)
        phase (float): phase shift in radians (default 0)

    Returns:
        array: sine wave evaluated at 'time'
    """
    return amplitude * sin(2 * pi * frequency * time + phase)
````

````python
import matplotlib.pyplot as plt
from numpy import linspace

# Time values
t = linspace(0, 2, 500)  # 0 to 2 seconds

# Default sine wave
y1 = func1(t)

# Higher amplitude
y2 = func1(t, amplitude=2)

# Higher frequency
y3 = func1(t, frequency=3)

# With phase shift (π/2 rad)
y4 = func1(t, phase=pi/2)

# Plot results
plt.figure(figsize=(10,6))
plt.plot(t, y1, label="Default (A=1, f=1Hz, φ=0)")
plt.plot(t, y2, label="Amplitude=2")
plt.plot(t, y3, label="Frequency=3 Hz")
plt.plot(t, y4, label="Phase=π/2")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Sine waves with different amplitude, frequency, and phase")
plt.legend()
plt.grid(True)
plt.show()
````

## Proof of Satellite Altitude Formula

We begin by equating the **gravitational force** acting on the satellite to the **centripetal force** required to keep it in circular motion.

### Step 1: Gravitational Force

\[
F_g = \frac{G M m}{r^2}
\]

### Step 2: Centripetal Force

\[
F_c = m \omega^2 r
\]

Equating both and simplifying:

\[
\frac{G M}{r^2} = \omega^2 r \quad \Rightarrow \quad G M = \omega^2 r^3
\]

Then substitute \( \omega = \frac{2\pi}{T} \):

\[
r = \left( \frac{G M T^2}{4\pi^2} \right)^{1/3}
\]

Finally, altitude above Earth’s surface:

\[
\boxed{h = \left( \frac{G M T^2}{4\pi^2} \right)^{1/3} - R}
\]

````python
# Constants (SI units)
G = 6.67430e-11       # Gravitational constant (m^3 kg^-1 s^-2)
M = 5.972e24          # Mass of Earth (kg)
R = 6.371e6           # Radius of Earth (m)

from math import pi

def satellite_altitude(T_seconds):
    """
    Calculate the altitude (h) of a satellite in a circular orbit around Earth.

    Parameters:
        T_seconds (float): Orbital period in seconds.

    Returns:
        float: Altitude h in meters above Earth's surface.
    """
    orbital_radius_cubed = (G * M * T_seconds**2) / (4 * pi**2)
    orbital_radius = orbital_radius_cubed**(1/3)
    h = orbital_radius - R
    return h
````

  1. [4] Write a function `linfit(x,y)` to perform a least squares fit of data using the method described last year in PX1224 week 4 to return estimates of slope and intercept (and their errors).
  2. [2] Put your functions into your own package, using a suitable file name. Type the code you want to be in the package (including comments!) in a cell of the notebook, pick a name for the file and add `%%writefile <file_name>.py` at the top of the cell to have the notebook create the file.

**IMPORTANT**: The `linfit` function itself must be self contained. It must only take in the x- and y-data, and return m, dm, c and dc. It must not do anything else, such as print or plot data - that should all be done outside the function.

````python
%%writefile linfittools.py

"""
linfittools.py

Contains the linfit(x, y) function which performs an unweighted least squares linear fit
to x-y data and returns slope, intercept, and their uncertainties.
"""

def linfit(x, y):
    """
    Perform an unweighted least squares linear  fit.

    Parameters:
        x (list or array): x data
        y (list or array): y data

    Returns:
        m  (float): slope of best-fit line
        dm (float): uncertainty in slope
        c  (float): y-intercept
        dc (float): uncertainty in intercept
    """
    import numpy as np

    x = np.array(x)
    y = np.array(y)

    N = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    S_xx = np.sum((x - x_mean)**2)
    S_xy = np.sum((x - x_mean)*(y - y_mean))

    m = S_xy / S_xx
    c = y_mean - m * x_mean

    # Residuals and variance
    y_fit = m * x + c
    residuals = y - y_fit
    variance = np.sum(residuals**2) / (N - 2)

    dm = (variance / S_xx) ** 0.5
    dc = (variance * (1/N + (x_mean**2)/S_xx)) ** 0.5

    return m, dm, c, dc
````

````python
# Import the function from your saved package
from linfittools import linfit
import numpy as np
import matplotlib.pyplot as plt

# Define test data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.1, 4.0, 6.1, 8.0, 10.2])

# Perform the fit
m, dm, c, dc = linfit(x_data, y_data)

# Print the results
print(f"Slope (m)     = {m:.3f} ± {dm:.3f}")
print(f"Intercept (c) = {c:.3f} ± {dc:.3f}")

# Generate fitted line
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = m * x_fit + c

# Plot
plt.figure(figsize=(7, 5))
plt.plot(x_data, y_data, 'o', label='Data points')
plt.plot(x_fit, y_fit, '-', label='Best fit line')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Squares Linear Fit")
plt.legend()
plt.grid(True)
plt.show()
````
